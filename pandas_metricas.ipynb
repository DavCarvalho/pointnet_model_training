Interpretação das Métricas:

- Overall Accuracy (OA): Representa a proporção de previsões corretas em relação ao total de previsões.

- Mean Intersection over Union (mIoU): Mede a precisão da segmentação calculando a interseção e união entre as previsões e os rótulos reais.

- Perda (Loss): Função que mede a diferença entre as previsões do modelo e os rótulos reais, usada para treinar o modelo ajustando os pesos.
